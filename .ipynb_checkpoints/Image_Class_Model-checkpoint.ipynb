```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf

from tensorflow import keras
from tensorflow.keras import layers
```

```python
data_train_path = 'solid_dry_waste/train'
data_test_path = 'solid_dry_waste/test'
data_val_path = 'solid_dry_waste/validation'
```

```python
img_width = 180
img_height =180 
```

```python
data_train = tf.keras.utils.image_dataset_from_directory(
    r'C:\Users\James\OneDrive\Desktop\EcoMend\solid_dry_waste\train',  # Corrected the path to raw string
    shuffle=True,
    image_size=(img_width, img_height),
    batch_size=32,
    validation_split=False
)


```

```python
data_cat = data_train.class_names
```

```python
data_cat
```

```python
data_val_path = r'C:\Users\James\OneDrive\Desktop\EcoMend\solid_dry_waste\validation'  # Ensure this path is correct

data_val = tf.keras.utils.image_dataset_from_directory(
    data_val_path,
    image_size=(img_height, img_width),
    batch_size=32,
    shuffle=False,
    validation_split=False
)
```

```python
data_val = tf.keras.utils.image_dataset_from_directory(
    data_val_path,
    image_size=(img_height, img_width),
    batch_size=32,
    shuffle=False
)
```

```python
plt.figure(figsize=(10,10))
for image, labels in data_train.take(1):
    for i in range(9):
        plt.subplot(3,3,i+1)
        plt.imshow(image[i].numpy().astype('uint8'))
        plt.title(data_cat[labels[i]])
        plt.axis('off')
```

```python
from tensorflow.keras.models import Sequential
```

```python
data_train
```

```python
model = Sequential([
    layers.Rescaling(1./255),
    layers.Conv2D(16, 3, padding='same', activation='relu'),
    layers.MaxPooling2D(),
    layers.Conv2D(32,3, padding='same',activation='relu'),
    layers.MaxPooling2D(),
    layers.Conv2D(64, 3, padding='same', activation='relu'),
    layers.MaxPooling2D(),
    layers.Flatten(),
    layers.Dropout(0.2),
    layers.Dense(128),
    layers.Dense(len(data_cat))
                  
])
```

```python
model.compile(optimizer='adam', loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True), metrics=['accuracy'])
```

```python
epochs_size = 25
history = model.fit(data_train, validation_data=data_val, epochs=epochs_size)
```

```python
epochs_range = range(epochs_size)
plt.figure(figsize=(8,8))
plt.subplot(1,2,1)
plt.plot(epochs_range,history.history['accuracy'],label = 'Training Accuracy')
plt.plot(epochs_range, history.history['val_accuracy'],label = 'Validation Accuracy')
plt.title('Accuracy')

plt.subplot(1,2,2)
plt.plot(epochs_range,history.history['loss'],label = 'Training Loss')
plt.plot(epochs_range, history.history['val_loss'],label = 'Validation Loss')
plt.title('Loss')
```

```python
# Full path to the image
image_path = r'C:\Users\James\OneDrive\Desktop\EcoMend\plastic.jpg'

# Load the image
image = tf.keras.utils.load_img(image_path, target_size=(img_height, img_width))

# Convert the loaded image to an array
img_arr = tf.keras.utils.img_to_array(image)

# Create a batch dimension
img_bat = tf.expand_dims(img_arr, axis=0)
```

```python
predict = model.predict(img_bat)
```

```python
score = tf.nn.softmax(predict)
```

```python
print('Veg/Fruit in image is {} with accuracy of {:0.2f}'.format(data_cat[np.argmax(score)],np.max(score)*100))
```
